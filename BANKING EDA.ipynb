```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python



df = pd.read_csv('Banking.csv')
df.head(5)
```

```python
df.shape
```

```python
df['Estimated Income'].value_counts()
```

```python
df['Estimated Income'].min()
```

```python
bins=[0,100000,300000,float('inf')]
labels=['low','medium','high']

df['Income band']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df['Income band'].value_counts()
```

```python
df['Income band'].value_counts().plot(kind='bar')
```

```python
df['Income band'].value_counts().plot()
```

```python
categoriccal_cols = df[
    [
        "BRId",                     # not BRID
        "GenderId",                 # not GenderID
        "IAId",
        "Amount of Credit Cards",
        "Nationality",
        "Occupation",
        "Fee Structure",
        "Loyalty Classification",
        "Properties Owned",
        "Risk Weighting",
        "Income band"               # not Income Band
    ]
].columns

for col in categoriccal_cols:
    print(col)
    display(df[col].value_counts())
    
```

UNIVARIATE ANALYSIS



```python
for i,predictor in enumerate(categoriccal_cols):
    plt.figure(i)
    sns.countplot(x=predictor,data=df)

#x=predictor tells seaborn which column name to use on x-axis.
#data=df tells seaborn where that column exists (inside df).
#So seaborn reads it like:
#“Take column predictor from DataFrame df and plot value counts.”

```

BIVARIATE ANALYSIS

```python
for i,predictor in enumerate(categoriccal_cols):
    plt.figure(i)
    sns.countplot(x=predictor,data=df,hue="Nationality")
```

HISTOGRAM OF OCCUPATION


```python
for col in categoriccal_cols:
    if col =="occupation":
     continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title("Histogram of occupation")
```

NUMERICAL ANALYSIS
UNIVARIATE


```python
numerical_cols=['Estimated Income','Superannuation Savings']
plt.figure(figsize=(10,10))
for i,col in enumerate(numerical_cols):
    plt.subplot(2,1,i+1)
    sns.histplot(df[col],kde=True)
    
plt.show()
```

HEATMAPS

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title("CM")
plt.show()
```
